```
imR = Image.open("/content/drive/MyDrive/RGBCMY.jpeg") 
h = int(imR.height)
w = int(imR.width)

for y in range(h):
  for x in range(w):
    rgb_im = imR.convert('RGB')
    #rgb_im = im2.convert('RGB')
    r, g ,b = rgb_im.getpixel((x, y))
    imR.putpixel((x, y), (r, 0, 0, 0))

plt.imshow(imR)
```
